```python
from SIRStochasticDynamics import *
from SISStochasticDynamics import *
from SIRStochasticDynamicsDisconnect import *
from SISStochasticDynamicsDisconnect import *
from SIRStochasticDynamicsRewire import *
from SISStochasticDynamicsRewire import *
```

STOCHASTIC

Part 1 - SIR (no rewire)

```python
# disease parameters
p_infected = 0.01
p_infect = 0.001
p_recover = 0.001

# network parameters
N = 5000
M = 3

# simulation parameters
time_limit = 20000

sto_SIR = SIRStochasticDynamics(time_limit = time_limit, p_infected = p_infected, p_infect = p_infect, p_recover = p_recover,
                             graph = networkx.barabasi_albert_graph(N,M))

sto_dyn_SIR = sto_SIR.dynamics()
```

```python
sto_SIR.plot_distributions('Stochastic - SIR - no rewire')
```
